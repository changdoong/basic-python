## 컬럼 기준 데이터 합치기

```python
import pandas as pd

employee_df = pd.read_csv('employee.csv')
survey_df = pd.read_csv('survey.csv')
```

```python
employee_df
```

```python
survey_df
```

```python
pd.merge(employee_df,survey_df, on='id')
```

### left outer 조인 하는법

```python
pd.merge(employee_df,survey_df, on='id',how='left')
```

### right outer 조인 하는법

```python
pd.merge(employee_df,survey_df, on='id',how='right')
```

### full outer 조인 하는법

```python
pd.merge(employee_df,survey_df, on='id',how='outer') # 자동적으로 department_x,_y가 생김
```

```python
pd.merge(employee_df,survey_df, on='id',how='outer',suffixes=('','_x'))
```

```python
survey_df = survey_df.rename(columns={'id': 'employee_id'})
survey_df
```

```python
pd.merge(employee_df, survey_df, left_on='id',right_on='employee_id')
```

```python
company_df = pd.merge(employee_df, survey_df, left_on='id',right_on='employee_id')
```

```python
company_df
```

## 인덱스 기준 데이터 합치기

```python
employee_df = employee_df.set_index('id')
survey_df = survey_df.set_index('employee_id')
```

```python
employee_df.join(survey_df) # suffix 오류
```

```python
employee_df.join(survey_df, lsuffix='_x', rsuffix='_y')
```

```python
employee_df.join(survey_df, rsuffix='_y')# 기본적으로 left outer join
```

```python
employee_df.join(survey_df, rsuffix='_y',how = 'inner')
```

```python
company_df = employee_df.join(survey_df, rsuffix='_y',how = 'inner')
```

```python
company_df
```

```python
condition = company_df['resign'] =='Y'
```

```python
company_df[condition]
```

## 해고당한 사람과 그렇지 않은 사람의 만족도 비교

```python
company_df[condition].describe()
```

```python
company_df[~condition].describe()
```
