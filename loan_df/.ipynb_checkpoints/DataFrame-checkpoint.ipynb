## 불린 인덱싱하기

```python
loan_df
```

```python
condition1 = loan_df['married'] == 'Y'
```

```python
loan_df['income'].mean()
```

```python
condition2 = loan_df['income'] > loan_df['income'].mean()
```

```python
loan_df[condition1 & condition2]
```

```python
loan_df[condition1 | condition2]
```

```python
loan_df[(loan_df['married']=='Y') | (loan_df['income'] > loan_df['income'].mean())]
```

#### 결혼을 했거나, 수입이 전체 평균보다 크고, 부동산이 도시에 있지 않다

```python
condition3 = loan_df['property_area'] == 'City'
```

```python
loan_df[condition1 | condition2 & ~condition3]
```

### query() 코드가 좀 더 간단해짐

```python
loan_df.query('income > 5000')
```

```python
income_mean = loan_df['income'].mean()
```

```python
loan_df.query('income > @income_mean')
```

```python
loan_df.query("property_area == 'City'")
```

```python
loan_df.query("property_area in ['City','Suburb']")
```

```python
loan_df.query("married == 'Y' & income > @income_mean")
```
